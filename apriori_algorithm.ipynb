#  Market Basket Analysis using Apriori Algorithm

###  Objective:
To find frequent itemsets and association rules from transaction data.

###  Algorithm Used:
- Apriori Algorithm

###  Dataset:
Groceries Dataset (Kaggle)

##  Step 1: Import Libraries

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

##  Step 2: Load Dataset

```python
import zipfile
zip_path = "dataset.zip"

# Open the zip file
with zipfile.ZipFile(zip_path, 'r') as z:
    # List files inside zip
    print(z.namelist())
    
    # Replace 'your_file.csv' with actual file name inside zip
    with z.open('Groceries_dataset.csv') as f:
        df = pd.read_csv(f)

# Print first 5 rows
print(df.head())
```

##  Step 3: Data Preprocessing

Convert data into transaction format

```python
# Convert dataset into transactions
transactions = df.groupby('Member_number')['itemDescription'].apply(list)

# Convert to list format
transactions = transactions.tolist()

# Show sample transactions
transactions[:5]
```

## Step 4: One-Hot Encoding

Convert transactions into binary matrix

```python
te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)

df = pd.DataFrame(te_array, columns=te.columns_)

df.head()
```

##  Step 5: Apply Apriori Algorithm

Generate frequent itemsets

```python
frequent_itemsets = apriori(df, min_support=0.02, use_colnames=True)

# Show top itemsets
frequent_itemsets.sort_values(by="support", ascending=False).head(10)
```

## Step 6: Generate Association Rules

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)

# Sort by lift
rules = rules.sort_values(by="lift", ascending=False)

rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']].head(10)
```

##  Step 7: Filter Strong Rules

High confidence and high lift rules

```python
strong_rules = rules[
    (rules['confidence'] >= 0.6) &
    (rules['lift'] > 1)
]

strong_rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
```

##  Step 8: Save Results

```python
frequent_itemsets.to_csv("frequent_itemsets.csv", index=False)
rules.to_csv("association_rules.csv", index=False)

print(" Results saved successfully!")
```

```python
def recommend_products_smart(rules, user_items):
    recommendations = {}
    
    user_items = [item.strip().lower() for item in user_items]

    for _, row in rules.iterrows():
        antecedent = [item.strip().lower() for item in row['antecedents']]
        consequent = list(row['consequents'])
        confidence = row['confidence']

        # ✅ Agar user item antecedent me hai (partial match)
        if any(item in antecedent for item in user_items):
            for item in consequent:
                if item not in user_items:
                    if item not in recommendations:
                        recommendations[item] = confidence
                    else:
                        recommendations[item] += confidence

    # Sort by confidence (best recommendations first)
    sorted_recommendations = sorted(recommendations.items(), key=lambda x: x[1], reverse=True)

    return [item[0] for item in sorted_recommendations]
```

```python
user_input = ['whole milk']

recommended = recommend_products_smart(rules, user_input)

print("Recommended:", recommended[:10])
```

```python
recommended = recommend_products_smart(rules, ['whole milk'])

print("Top 5 Recommendations:", recommended[:5])
```

```python
def recommend_with_score(rules, user_items):
    results = {}

    user_items = [item.strip().lower() for item in user_items]

    for _, row in rules.iterrows():
        antecedent = [i.strip().lower() for i in row['antecedents']]
        consequent = list(row['consequents'])
        confidence = row['confidence']

        if any(item in antecedent for item in user_items):
            for item in consequent:
                if item not in results:
                    results[item] = confidence
                else:
                    results[item] += confidence

    return sorted(results.items(), key=lambda x: x[1], reverse=True)


print(recommend_with_score(rules, ['whole milk']))
```

```python
# Only strong rules (better results)
strong_rules = rules[(rules['confidence'] > 0.3) & (rules['lift'] > 1)]

# Store combinations
combinations = []

for _, row in strong_rules.iterrows():
    items = list(row['antecedents']) + list(row['consequents'])
    
    if len(items) >= 2:
        combinations.append({
            "items": items,
            "confidence": row['confidence'],
            "lift": row['lift']
        })

# Sort by lift (strong relation first)
combinations = sorted(combinations, key=lambda x: x['lift'], reverse=True)

# Show top combinations
for combo in combinations[:10]:
    print(combo)
```
